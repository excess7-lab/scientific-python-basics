# ACTIVIDADES QUE SE REALIZARON
En el presente notebook, primero se importaron las librerías necesarias, que en este caso son numpy y matplotlib. 
Luego se crearon los arrays, linspace retorna un arreglo de números igualmente espaciados, en este caso, el primer parámetro es el punto desde donde se comenzará a contar, el segundo es hasta donde se llegará y el tercero nos indica cuantos número habrán entre ellos. 
El siguiente random.randn retorna una muestra desde una distribución estándar, los parámetros son las dimensiones del arreglo que retornará y deben ser positivos, si no hay parámetros, se retorna un punto flotante. 
numpy.eye retorna un arreglo bidimensional de unos en la diagonal y ceros en otros lugares, el parámetro ingresado indica el número de filas en la salida. 
Las siguientes operaciones np.mean() devuelve el valor medio del arreglo v, std la desviacióon estándar y v**2 me devuelve un arreglo de la misma forma, pero elevado al cuuadrado cada valor. Luego sum sumará los elementos de un arreglo sobre un eje seleccionado, en este caso, el eje cero.
Por último se simularon 1000 ingresos mensuales,  con una media de 800 y un valor de desviación estándar de 120.
